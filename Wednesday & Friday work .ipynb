## Question 6. 
(Hannah helped me with the how to aggregate it)
Build a visualization that allows you to explore the correlation of word counts for Marlowe and Shakespeare. What does this tell you about the styles of the two authors? Note: You will need to aggregate your data to answer this question.

```python
alt.Chart(rawCount, width = 300).mark_bar().encode(
    x = "Author",
    y = alt.Y(field='Count', aggregate='mean', type='quantitative')
)
```

```python
alt.Chart(rd, width = 300).mark_point().encode(
    x = "Author",
    y = "Negative"
).interactive()



```

The graph aboveshows that although very close, Shakespear has more negative sentiment within his writtings due to the larger distribution of dots within the graph. However is it a very close tie between the two. 

```python
alt.Chart(rd, width = 100).mark_bar().encode(
    x='Author',
    y='Text',
    color='Author'

)

```

This graph above is interesting to see because this shows that Shakespear had more writtings overall, with the larger amount of orange bars. 

```python
alt.Chart(rd, width = 500).mark_bar().encode(
    x='Text',
    y='Positive',
    color= 'Author'

)
```

The chart above shows the correlation between the texts and their positive sentiment as well as divides the text by color of the author. This tells you that Marlowe had a text that had the highest posotive sentiment, but Shakespear wrote more text and those text were slightly more positive than Marlowes texts.
